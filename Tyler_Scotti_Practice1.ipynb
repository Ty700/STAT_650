### Task 1: Merge

1. Create two DataFrames:
    - df1 with columns 'ID', 'Name', 'Age' and values [(1, 'Alice', 25), (2, 'Bob', 30), (3, 'Charlie', 22)].
    - df2 with columns 'ID', 'City' and values [(1, 'New York'), (2, 'San Francisco'), (4, 'Chicago')].
2. Merge df1 and df2 using the common column 'ID'. Perform an inner join.

```python
import pandas as pd

df1 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 22]
})

df2 = pd.DataFrame({
    'ID': [1, 2, 4],
    'City': ['New York', 'San Francisco', 'Chicago']
})

merged_df = pd.merge(df1, df2, on='ID', how='inner')
print(merged_df)
```

### Task 2: Join

1. Create two DataFrames:
    - df3 with columns 'ID', 'Salary' and values [(1, 60000), (2, 80000), (3, 55000)].
    - df4 with columns 'ID', 'Bonus' and values [(1, 5000), (2, 7000), (4, 4000)].
2. Join df3 and df4 using the common column 'ID'. Perform a left join.

```python
df3 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Salary': [60000, 80000, 55000]
})

df4 = pd.DataFrame({
    'ID': [1, 2, 4],
    'Bonus': [5000, 7000, 4000]
})

joined_df = pd.merge(df3, df4, on='ID', how='left')
print(joined_df)
```

### Task 3: Concatenate

1. Create two DataFrames:
    - df5 with columns 'ID', 'Subject' and values [(1, 'Math'), (2, 'Physics'), (3, 'Chemistry')].
    - df6 with columns 'ID', 'Subject' and values [(4, 'Biology'), (5, 'History'), (6, 'English')].
2. Concatenate df5 and df6 along the rows.

```python
df5 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Subject': ['Math', 'Physics', 'Chemistry']
})

df6 = pd.DataFrame({
    'ID': [4, 5, 6],
    'Subject': ['Biology', 'History', 'English']
})

concatenated_df = pd.concat([df5, df6], ignore_index=True)
print(concatenated_df)
```

### Task 4: DataFrame Creation and Operations

Create a Pandas DataFrame named **df** with the following data:

1. Display the first 3 rows of the DataFrame.
2. Calculate and display the average age of individuals in the DataFrame.
3. Add a new column named 'Salary' with values [60000, 80000, 55000, 75000, 70000].

```python
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [25, 30, 22, 35, 28],
    'City': ['New York', 'San Francisco', 'Los Angeles', 'Chicago', 'Miami']
}

df = pd.DataFrame(data)

print("First 3 rows:")
print(df.head(3))

avg_age = df['Age'].mean()
print(f"\nAverage age: {avg_age}")

df['Salary'] = [60000, 80000, 55000, 75000, 70000]
print("\nDataFrame with Salary column:")
print(df)
```

### Task 5: Data Selection and Filtering

1. Select and display only the 'Name' and 'City' columns of the DataFrame.
2. Filter and display rows where the age is greater than 25.
3. Display the information summary of the DataFrame.

```python
print("Name and City columns:")
print(df[['Name', 'City']])

print("\nRows where Age > 25:")
print(df[df['Age'] > 25])

print("\nDataFrame info:")
print(df.info())
```

### Task 6: Data Aggregation

1. Calculate and display the maximum age in the DataFrame.
2. Group the DataFrame by the 'City' column and calculate the average age for each city.

```python
max_age = df['Age'].max()
print(f"Maximum age: {max_age}")

print("\nAverage age by city:")
avg_age_by_city = df.groupby('City')['Age'].mean()
print(avg_age_by_city)
```

### Task 7: Data Visualization

1. Plot a bar chart showing the average age for each city.

```python
import matplotlib.pyplot as plt

avg_age_by_city = df.groupby('City')['Age'].mean()

plt.figure(figsize=(10, 6))
avg_age_by_city.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Average Age by City', fontsize=14, fontweight='bold')
plt.xlabel('City', fontsize=12)
plt.ylabel('Average Age', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```
