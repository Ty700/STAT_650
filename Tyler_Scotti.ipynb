### Question 1
First, write a Python function that converts a temperature from Celsius to Fahrenheit and a function that converts from Fahrenheit to Celsius. Then, test your functions on the values of 25C and 77F.

```python
# Define two functions
def c_to_f(temp) -> float:
    return (temp * (9/5)) + 32


def f_to_c(temp) -> float:
    return (temp - 32) * (5/9)


# Test the functions
assert c_to_f(25) == 77, "c_to_f is not working"
assert f_to_c(77) == 25, "f_to_c is not working"
```

### Question 3
Write a Python program to generate a multiplication table (1 to 10) for numbers 1 to 5 using nested `for` loops.

```python
# Nested FOR loops
for i in range(1, 6):
  print(f"Multiplication table for {i}:")
  for j in range(1, 11):
    print(f"{i} * {j} = {i * j}")
  print()
```

### Question 4
Write a Python function that takes a list of integers and returns only the even numbers. Then test you function on the list: (1,2,3,4,5,6).

```python
# Define function
def filter_even(lst: list[int]) -> list[int]:
    return [x for x in lst if x % 2 == 0]

# Test the function and Print even numbers
assert filter_even([1, 2, 3, 4, 5, 6]) == [2, 4, 6], "filter_even is not working"
print(filter_even([1, 2, 3, 4, 5, 6]))
```

Use a `while` loop to print numbers from 10 down to 1. Use `-=`.

### Question 6
Use as your data the sequence of whole numbers from 1 to 10. Write a Python program that prints your data with a `break` at i=8 and, and a `continue` statement for i=5.

```python
# Your code here
seq = range(1, 11)

for i in seq:
    if i == 8:
        break
    elif i == 5:
        continue
    print(i)
```

### Question 7
Write a Python function that uses recursion to calculate the factorial of a number. Test your function with 5.

```python
# Define factorial function
def factorial(n: int) -> int:
  if n == 0:
    return 1
  else:
    return n * factorial(n - 1)

# Test the function on 5
assert(factorial(5) == 120), "factorial is not working"
print(factorial(5))
```
